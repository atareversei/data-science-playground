```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('../data/iris.csv')
df.sample(5)
```

```python
plt.scatter(df['sepal.length'], df['sepal.width'], c=df['variety'].map({'Setosa': 0, 'Versicolor': 1, 'Virginica': 2}), alpha=0.3)
plt.title('Sepal Length and Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
fig = px.scatter_3d(df,
                    x="sepal.length",
                    y="sepal.width",
                    z="petal.length",
                    color="variety",
                    title="Interactive 3D scatter plot of Iris Dataset",
                    labels={'variety': "Species"},
                    )

fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
fig.show()
```
